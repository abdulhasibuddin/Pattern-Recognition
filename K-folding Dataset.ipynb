```python
dataset = [i for i in range(101)]
print(dataset)
```

```python
total_data = len(dataset)
k = 10
split_indices = int(total_data/k)
```

```python
for fold in range(1,k+1):
    training_dataset = dataset[:fold*split_indices-split_indices]
    training_dataset += dataset[fold*split_indices:]
    testing_dataset = dataset[fold*split_indices-split_indices:fold*split_indices]

    print('For k=',fold,'\n')
    print('training_dataset =',training_dataset)
    print('\ntesting_dataset =',testing_dataset)
    print('---------------------------------\n')
```
